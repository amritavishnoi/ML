```python
#dropping the unnecessary columns
employee = emp.drop(['EmployeeCount','EmployeeNumber','Over18','StandardHours'],axis=1)
```

```python
employee.shape
```

```python
# Dummy variables
employee['Attrition']=np.where(employee['Attrition']=='Yes',1,0)
```

```python
employee.head()
```

```python
print(employee.BusinessTravel.value_counts())
print('***************************************')
print(employee.Department.value_counts())
print('***************************************')
print(employee.EducationField.value_counts())
print('***************************************')
print(employee.JobRole.value_counts())
print('***************************************')
```

```python
#encoding categorical variables into numerical variables
employee_dummies=pd.get_dummies(employee[['BusinessTravel','Department','EducationField','Gender',\
                                          'JobRole','MaritalStatus','OverTime']],drop_first=True)
```

```python
employee=employee.drop(['BusinessTravel','Department','EducationField','Gender',
                        'JobRole','MaritalStatus','OverTime'],axis=1)
```

```python
employee=pd.concat([employee,employee_dummies],axis=1)
```

```python
employee.head()
employee.shape
```

```python
x=employee.drop(['Attrition'],axis=1)
y=employee['Attrition']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=20)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train,y_train)

y_pred = gnb.predict(X_test)

```

```python
from sklearn import metrics

print("Accuracy: " ,metrics.accuracy_score(y_test,y_pred))
```
